```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv("/content/Unemployment in India.csv")
df2 = pd.read_csv("/content/Unemployment_Rate_upto_11_2020.csv")

df1.head(), df2.head()
```

```python
df1.isnull().sum()
df2.isnull().sum()

df1 = df1.dropna()
df2 = df2.dropna()
```

```python
df1.columns = df1.columns.str.strip()
df2.columns = df2.columns.str.strip()
```

```python
df1['Date'] = pd.to_datetime(df1['Date'], dayfirst=True)
df2['Date'] = pd.to_datetime(df2['Date'], dayfirst=True)
```

```python
plt.figure(figsize=(10,5))
plt.plot(df1['Date'], df1['Estimated Unemployment Rate (%)'])
plt.title("Unemployment Rate Trend in India")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.xticks(rotation=45)
plt.show()
```

```python
covid_data = df1[df1['Date'] >= '2020-03-01']

plt.figure(figsize=(10,5))
plt.plot(covid_data['Date'], covid_data['Estimated Unemployment Rate (%)'], color='red')
plt.title("Impact of Covid-19 on Unemployment in India")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(
    data=df1,
    x='Region',
    y='Estimated Unemployment Rate (%)'
)
plt.xticks(rotation=90)
plt.title("Region-wise Unemployment Rate")
plt.show()
```

```python
df1['Month'] = df1['Date'].dt.month

plt.figure(figsize=(8,5))
sns.boxplot(x='Month', y='Estimated Unemployment Rate (%)', data=df1)
plt.title("Seasonal Trend in Unemployment Rate")
plt.show()
```
